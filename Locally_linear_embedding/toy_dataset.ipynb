```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import offsetbox
from sklearn.manifold import LocallyLinearEmbedding
import sklearn.datasets as ds
```

```python
# first run lle/mlle on sample datasets
# reduce the dimensions to 2 or 3 and plot them
# gaussian with 3 quantiles
# few blobs
plt.figure(figsize=(8,8))

# 3 blobs

#---------------------------Standard method on 3blobs----------------------
X1,y1 = ds.make_blobs(n_samples=500,n_features=30)

LLE2d = LocallyLinearEmbedding(n_neighbors=30, n_components=2)

LLE2d.fit(X1)
#LLE3d.fit(X1)


plt.subplot(221)
plt.title('standard method on 3 blobs')
plt.scatter(LLE2d.embedding_[:,0],LLE2d.embedding_[:,1],c=y1)



#------------------------Standard method on gaussian------------

X2,y2 = ds.make_gaussian_quantiles(n_classes=3,n_features=30,n_samples=500)

#LLE3d = LocallyLinearEmbedding(n_neighbors=30, n_components=3)
LLE2d = LocallyLinearEmbedding(n_neighbors=30, n_components=2)

LLE2d.fit(X2)
#LLE3d.fit(X2)

plt.subplot(222)
plt.title('standard method on gaussian with 3 classes')
plt.scatter(LLE2d.embedding_[:,0],LLE2d.embedding_[:,1],c=y2)


#--------------standard method on iris ------------------

iris = ds.load_iris()
#print(iris.target.shape,iris.data.shape)
LLE = LocallyLinearEmbedding(n_neighbors=10,n_components=2)
LLE.fit(iris.data)

plt.subplot(223)
plt.title('standard method on iris data')
plt.scatter(LLE.embedding_[:,-0],LLE.embedding_[:,1],c=iris.target)

#-------------standard method on handwritten digits----------------

dig = ds.load_digits()
LLE = LocallyLinearEmbedding(n_components=2,n_neighbors=30)
LLE.fit(dig.data)

plt.subplot(224)
plt.title('standard method on digit data')
plt.scatter(LLE.embedding_[:,0],LLE.embedding_[:,1],c=dig.target)

```

```python
# 3D plotting with same data
#------------------standard method on 3blob------------
X,Ys = ds.make_blobs(n_samples=500,n_features=30)

LE = LocallyLinearEmbedding(n_components=3,n_neighbors=30)
LE.fit(X1)
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(2,2,1, projection='3d')
X=LE.embedding_[:,0]
Y = LE.embedding_[:,1]
Z = LE.embedding_[:,2]
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

ax.scatter(X,Y,Z,c=Ys)
#------------------standard method on gaussian------------
#LE = LocallyLinearEmbedding(n_components=3,n_neighbors=30)
x,y = ds.make_gaussian_quantiles(n_classes=3,n_features=50,n_samples=500)
LE = LocallyLinearEmbedding(n_components=3,n_neighbors=30)
LE.fit(x)

ax = fig.add_subplot(2,2,2,projection='3d')
X=LE.embedding_[:,0]
Y = LE.embedding_[:,1]
Z = LE.embedding_[:,2]

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

ax.scatter(X,Y,Z,c=y)
#------------------standard method on iris------------
iris = ds.load_iris()
#print(iris.target.shape,iris.data.shape)
LE = LocallyLinearEmbedding(n_components=3,n_neighbors=30)
LE.fit(iris.data)

ax = fig.add_subplot(2,2,3,projection='3d')
X=LE.embedding_[:,0]
Y = LE.embedding_[:,1]
Z = LE.embedding_[:,2]

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

ax.scatter(X,Y,Z,c=iris.target)
#------------------standard method on digits------------
LE = LocallyLinearEmbedding(n_components=3,n_neighbors=30)
dig = ds.load_digits()
LE.fit(dig.data)

ax = fig.add_subplot(2,2,4,projection='3d')
X=LE.embedding_[:,0]
Y = LE.embedding_[:,1]
Z = LE.embedding_[:,2]

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

ax.scatter(X,Y,Z,c=dig.target)


#=================================================#




plt.show()
```
