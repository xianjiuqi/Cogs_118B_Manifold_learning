```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import offsetbox
from sklearn.manifold import LocallyLinearEmbedding
```

```python
data = np.genfromtxt('hmnist_28_28_L.csv',delimiter=',')
X_and_Y = data[1:] #get rid of labels
X = X_and_Y[:,:-1]
y = X_and_Y[:,-1]
```

```python
#running standard and modified LLE on hmnist_28_28
stand = LocallyLinearEmbedding(n_components=2,n_neighbors=50,method='standard')
modi = LocallyLinearEmbedding(method='modified',n_components=2,n_neighbors=50)

stand.fit(X)
modi.fit(X)

fig = plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.scatter(stand.embedding_[:,0],stand.embedding_[:,1],c=y)
plt.title('standard method on colorectal data')

plt.subplot(1,2,2)
plt.scatter(modi.embedding_[:,0],modi.embedding_[:,1],c=y)
plt.title('modified method on colorectal data')
plt.show()
```

```python
# now 3d plotting
%matplotlib notebook
```

```python
stand = LocallyLinearEmbedding(n_components=3,n_neighbors=50, method ='standard')
modi = LocallyLinearEmbedding(n_components=3,n_neighbors=50, method ='modified')

stand.fit(X)
modi.fit(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,2,1,projection='3d')
X = stand.embedding_[:,0]
Y = stand.embedding_[:,1]
Z = stand.embedding_[:,2]

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

ax.scatter(X,Y,Z,c=y)

ax=fig.add_subplot(1,2,2, projection='3d')
X = modi.embedding_[:,0]
Y = modi.embedding_[:,1]
Z = modi.embedding_[:,2]
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

ax.scatter(X,Y,Z,c=y)
plt.show()
```
